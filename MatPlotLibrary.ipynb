```
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
import numpy as np
```

```
x = np.linspace(0,5,11)
y = x**2
```

```
x
```

```
y
```

**Functional Method**

```
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot of X vs. Y')
```

```
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

**Object Oriented Method**

```
fig = plt.figure()
#Axes have (left, bottom, width, height)
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel('X label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
axes.plot(x,y)
```

```
fig = plt.figure()
#Here we manually add the figure and below see we set it location and size
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
#Now dimesnions for different axis are set and then we manually do it
axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

Sub-plots With OOPS

```
fig,axes = plt.subplots(nrows=1, ncols=2)
#axes is a list so we will use it like that
#This method is setting axes according to rows and coloumns
axes[0].plot(x,y)
axes[0].set_title('First Plot')
#Here we get array of axes
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

#Remove Overlapping
plt.tight_layout()
```

**Figure-Size and DPI**

```
fig = plt.figure(figsize=(3,3)) #Adjust the size of the figure

fg = fig.add_axes([0,0,1,1])
fg.plot(x,y)
```

Sub-plots

```
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(6,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
#The axes will be treated as an array only when nrows are being initialed with greater number
plt.tight_layout()
```

Save the Figure

```
fig.savefig('my_picture.jpg', dpi=200)
```

**Adding Legend**

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X Squared')
ax.plot(x,x**3, label = 'Y Squared')

ax.legend(loc=0)
```

**Plot Appearence**

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
#Alpha is for changing transparency
ax.plot(x,y, color='red', linewidth=3, alpha=0.5, linestyle='--')
#for line style we can pass dashes steps 
```

Adding Markers

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#Adding the points on the line
ax.plot(y,x, color='red', marker='o',markersize=20, markerfacecolor='green', markeredgewidth=3, markeredgecolor='yellow')
```

Axis

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#Adding x and y limit points
ax.plot(y,x, color='red')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
