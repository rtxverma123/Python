```
import seaborn as sns
```

```
tips = sns.load_dataset('tips')
```

```
tips.head()
```

**Distribution Plot**

```
sns.displot(tips['total_bill'],bins=40) #How the particular value is being distributed
```

```
sns.distplot(tips['total_bill'])
```

**Joining Two Plots**

```
sns.jointplot(x='total_bill', y='tip', data=tips)
```

**Pair-Plot**
#It creates graph of all the index with each other 

```
sns.pairplot(tips)
```

```
sns.pairplot(tips,hue='sex')
```

Rug-plot

```
sns.rugplot(tips['total_bill'])
```
